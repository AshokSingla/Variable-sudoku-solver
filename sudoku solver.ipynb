```python
while True:
    n = int(input("Enter the size of sudoku : 4 for 4X4, 9 for 9X9 and 16 for 16X16.\n"))
    if n in [4,9,16]:
        break
    else:
        print("Invalid entry.\n\n")
    
sudoku = input("Enter Soduku")
# Assuming conventional 9X9 sudoku

# The correct way to enter is 070000009510420600080300700008001370023080040400900100962800030000010400700203096
# for the below sudoku.

#              0 7 0 0 0 0 0 0 9
#              5 1 0 4 2 0 6 0 0
#              0 8 0 3 0 0 7 0 0
#              0 0 8 0 0 1 3 7 0
#              0 2 3 0 8 0 0 4 0 
#              4 0 0 9 0 0 1 0 0
#              9 6 2 8 0 0 0 3 0
#              0 0 0 0 1 0 4 0 0 
#              7 0 0 2 0 3 0 9 6

puz = [ [int(sudoku[(i+j)]) for i in range(n)] for j in range(0,n*n,n)]

# Another way is to give the row-col command and make the changes.      # will implement in future

def check(puzzle, i, row, col,n):
    rows = puzzle[int(row)]
    column = [puzzle[r][col] for r in range(0,n)]
    if i in rows:
        return False
    if i in column:
        return False
    SquareRow = (row // int(n**.5))*int(n**.5)
    squareColumn = (col // int(n**.5))*int(n**.5)
    Square = [puzzle[y][z] for y in range(SquareRow, SquareRow+int(n**.5)) for z in range(squareColumn, squareColumn+int(n**.5))]
    if i in Square:
        return False
    return True


def find(puzzle,n):
    for i in range(0,n):
        for j in range(0,n):
            if puzzle[i][j]==0:
                return i,j
    return None


def solve(board,n):
    finds = find(board,n)
    if not finds:
        return True
    else:
        row, col = finds

    for i in range(1,n+1):
        if check(board, i, row, col,n):
            board[row][col] = i

            if solve(board,n):
                return True

            board[row][col] = 0

    return False

print(solve(puz,n))
for _ in puz:
    print(_)
```
